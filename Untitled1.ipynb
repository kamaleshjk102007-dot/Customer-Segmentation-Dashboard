```python
df = pd.read_excel('Sales_sample_data.xlsx')
```

```python
df.head()
```

```python
print(df.shape)
print(df.columns)
print(df.info())
```

```python
customer_df = df.groupby('customer').agg({
    'revenue':'sum',
    'profit':'sum',
    'quantity':'sum',
    'order_id':'count'
}).reset_index()
```

```python
customer_df.columns = [
    'customer',
    'total_revenue',
    'total_profit',
    'total_quantity',
    'total_orders'
]
```

```python
customer_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

features = [
    'total_revenue',
    'total_profit',
    'total_quantity',
    'total_orders'
]

X = customer_df[features]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range(1,11):
    model = KMeans(
        n_clusters=i,
        random_state=42,
        n_init=10
    )

    model.fit(X_scaled)

    wcss.append(model.inertia_)

plt.plot(range(1,11),wcss,marker='o')
plt.xlabel("Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=customer_df,
    x='total_revenue',
    y='total_profit',
    hue='Segment_Name',
    s=100
)

plt.title('Customer Segmentation using K-Means Clustering')
plt.xlabel('Total Revenue')
plt.ylabel('Total Profit')
plt.legend(title='Customer Segment')

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=4,
    random_state=42,
    n_init=10
)

customer_df['Segment'] = kmeans.fit_predict(X_scaled)
```

```python
customer_df.groupby('Segment')[['total_revenue', 'total_profit', 'total_quantity', 'total_orders']].mean()
```

```python
customer_df.groupby('Segment')[[
    'total_revenue',
    'total_profit',
    'total_quantity',
    'total_orders'
]].mean()
```

```python
segment_stats = customer_df.groupby('Segment').mean(numeric_only=True)

print(segment_stats)
```

```python
segment_names = {
    1: "VIP Customers",
    3: "High Value Customers",
    0: "Regular Customers",
    2: "Low Value Customers"
}

customer_df["Segment_Name"] = customer_df["Segment"].map(segment_names)
```
